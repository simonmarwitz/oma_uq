```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
pg=sns.pairplot(penguins, corner=True, diag_kind='scatter', hue="species")
axes=pg.axes
for i in range(axes.shape[0]):
    axes[i,i].twinx().annotate('abc',(0,0))
```
