```python
import numpy as np
import matplotlib.pyplot as plt
N=2**16

# system input (white noise)
phase=np.linspace(0,np.pi,N//2+1)
phase=(np.random.rand(N//2+1)-0.5)
Pomega = 10000*np.ones_like(phase)*np.exp(1j*phase*2*np.pi)

sig = np.fft.irfft(Pomega)
sig = np.random.randn(N)/np.sqrt(N)*10000

plt.figure()
plt.plot(sig)

plt.figure()
fft = np.abs(np.fft.rfft(sig))
plt.plot(fft)
plt.axhline(np.mean(fft))

plt.plot(np.abs(Pomega))
```
