```python
from sympy import *
init_printing()
k,kl,knl,x,a,dmax = symbols('k k_{\\text{l}} k_{\\text{nl}} x \\alpha d_{\\text{max}}')
```

```python
display('Equation for a cubic hardening/softening spring, consisting of a linear k_l and a nonlinear k_nl spring, where alpha denotes the fraction of linear to nonlinear stiffness parts')
spring = (1-a)*kl*x+a*knl*x**3
display(Eq(k,spring))
display('Solving for k_nl such that nonlinear spring force at maximal displacement d_max is always equal to a linear spring force')
knldmax = solve((spring-kl*x).subs(x,dmax),knl)[0]
display(Eq(knl,knldmax))
display('Substitutung k_nl into nonlinear spring equation')
spring3 = spring.subs(knl,knldmax)
display(Eq(k,spring3))
display('Compute minimal alpha, such that monotone stiffness is guaranteed.')
display(a>=solve(diff(spring3,x).subs(x,dmax),a)[0])
display('Plot of force-displacement relationship for alpha= [-0.5... 0.5],kl=10 and dmax=10')
dmaxval=20
klval=10
plot(*[spring3.subs(dmax,dmaxval).subs(kl,klval).subs(a,(i-10)/10) for i in range(5,16)],(x,0,dmaxval))
```

```python
d,omega,omegan, s,R,k1,zeta,m = symbols('d \\omega \\omega_n \\hat{s} R k_1 \\zeta m')
Wdf = 4*R*(s-R/k1) #Petersen Gl. 630
#Wdf = 4*R*s # Chopra 3.11.3
Wdv = d*pi*omega*s**2 # Petersen Gl. 555
#Wdv=2*pi*zeta*omega/omegan*k1*s**2 # Chopra 3.8.1
```

```python
display(solve(Eq(Wdv,Wdf),zeta))
display(solve(Eq(Wdv,Wdf),R))
```
